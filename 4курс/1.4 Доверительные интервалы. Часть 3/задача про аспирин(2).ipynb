```python
a_data = np.zeros(sample1)
p_data = np.zeros(sample2)
```

```python
a_data[0:(sample1 - 104)] = 1
p_data[0:(sample2 - 189)] = 1
```

```python
print len(a_data[a_data == 0])
print len(p_data[p_data == 0])
```

```python
def get_bootstrap_samples(data, n_samples):
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    samples = data[indices]
    return samples
```

```python
def stat_intervals(stat, alpha):
    #boundaries = np.percentile(stat, [100 - alpha / 2., 100 * (1 - alpha / 2.)])
    boundaries = np.percentile(stat, [])
    return boundaries
```

```python
np.random.seed(0)

a_meanScores = map(np.mean, get_bootstrap_samples(a_data, 1000))
p_meanScores = map(np.mean, get_bootstrap_samples(p_data, 1000))

print stat_intervals(a_meanScores, 0.05), '- a'
print stat_intervals(p_meanScores, 0.05), '- p'
```

```python
#Bp1 = p1/(1-p1)
#Bp2 = p2/(1-p2)
ber_mean_scores = map(lambda x: x[1] / x[0], zip(a_meanScores,p_meanScores))
print stat_intervals(ber_mean_scores, 0.05)
```
