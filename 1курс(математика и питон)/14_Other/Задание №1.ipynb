```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
b = 0.3
rv = sts.bradford(b)
```

```python
sample = rv.rvs(1000)
```

```python
x = np.linspace(rv.ppf(0.01),rv.ppf(0.99),100)
pdf = rv.pdf(x)
plt.plot(x, pdf)
plt.hist(sample, density=True)
plt.ylabel('$y$')
plt.xlabel('$x$')
```

```python
ex = rv.expect()
var = rv.var()
print (var)
```

```python
a = [np.average(rv.rvs(size=5)) for i in range(1000)]
norm_rv = sts.norm(ex, pow((var /5.0),0.5))
```
