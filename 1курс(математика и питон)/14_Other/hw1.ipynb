```python
# Базовые импорты
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# Пусть будет гамма распределение
coeff = 1.4
arr_random = sts.gamma.rvs(coeff, scale=1, size=1000) 
average = coeff # мат ожидание гамма распределения (= коэфф Гамма распределения)
variance = coeff**2 # дисперсия гамма распределения (= коэфф**2 Гамма распределения)
```

### Нарисуем гистограмму выборки и график теоретического распределения

```python
x = np.linspace(0,6,100)
#Гистограмма
plt.hist(arr_random, bins=10, density=True, color='xkcd:wheat')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
#График плотности гамма распределения
pdf = sts.gamma.pdf(x, coeff, scale=1)
plt.plot(x, pdf, label='gamma pdf', lw=5, alpha=0.5, color='xkcd:blue') #alpha - степень прозрачности
plt.legend()
```

### Перейдем к выборкам разного объема для выборочных средних

```python
# подготовим распределения выборочных средних для разных объемов (5, 10, 50)
mids5, mids10, mids50 = [], [], [] # list-ы хранящие средние для разных объемов (5, 10, 50)
for i in range(1000):
    arr_random = sts.gamma.rvs(coeff, size=5)   
    mids5.append( sum(arr_random)/5)
    arr_random = sts.gamma.rvs(coeff, size=10)   
    mids10.append( sum(arr_random)/10)
    arr_random = sts.gamma.rvs(coeff, size=50)   
    mids50.append( sum(arr_random)/50)

# подсчитаем стандартные отклонения для разных объемов (5, 10, 50)
stDev5, stDev10, stDev50 = math.sqrt(variance/5), math.sqrt(variance/10), math.sqrt(variance/50) # стандартное отклонение
```

```python
# будем работать с нормальными распределения, в качестве аргументов мат ожидание и подсчитанное ранее стандартное отклонение
norm5, norm10, norm50 = sts.norm(average, stDev5), sts.norm(average, stDev10), sts.norm(average, stDev50)
# и соответственные плотности
pdf5, pdf10, pdf50 = norm5.pdf(x), norm10.pdf(x), norm50.pdf(x)
```

### Будем визуализировать и смотреть как ЦПТ "улучшается" при увеличении объема выборки

```python
#Гистограмма и график для n = 5
plt.hist(mids5, bins=30, density=True, color='xkcd:red', label='n=3', alpha=0.3)
plt.plot(x, pdf5, label='n=5 pdf', color='xkcd:red', lw=5)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

```python
#Гистограмма и график для n = 10
plt.hist(mids10, bins=30, density=True, color='xkcd:green', label='n=10', alpha=0.3)
plt.plot(x, pdf10, label='n=10 pdf', color='xkcd:green', lw=5)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

```python
#Гистограмма и график для n = 50
plt.hist(mids50, bins=10, density=True, color='xkcd:blue', label='n=50', alpha=0.3)
plt.plot(x, pdf50, label='n=50 pdf', color='xkcd:blue', lw=5)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.legend()
```

### Выводы
1. Заметим что с ростом числа объема выборки изначальная "скошенность влево" графика распределения плотности выравнивается
2. С ростом числа объема выборки также исчезают всплески и гистограмма становится более гладкой, график уже более точно аппроксимизирует гистограмму
3. Для данного распределения переход от n=3 до n=50 уже хорошо демонстрирует работу ЦПТ.
