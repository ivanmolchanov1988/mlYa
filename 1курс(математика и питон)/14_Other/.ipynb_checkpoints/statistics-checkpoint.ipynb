```python
from scipy.stats import lognorm, norm
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from itertools import repeat
import math
```

Будет использоваться логнормальное распределение с параметром s=0.75 и $\mu$=21.0

```python
s = 0.75
mu = 21.0
f = lognorm(s, loc=mu)
sample = f.rvs(size=1000)
```

Теперь сравним графики теоретической плотности вероятности и полученной на основе выборки - они примерно равны.

```python
x = np.linspace(f.ppf(0.01), f.ppf(0.99), 100)
plt.plot(x, f.pdf(x))
plt.hist(sample, density=True, bins=40)
```

Делаем тысячу выборок по 3, 10, 50 и 200 элементов

```python
sample_means_by_n = dict()
ns = [3,10,50,200]
for n in ns:
    sample_means = list(
        map(lambda _: np.mean(f.rvs(size=n)), 
            repeat(None, 1000))
    )
    sample_means_by_n[n] = sample_means
```

Для сравнения нам интересно знать ожидаемые параметры распределения выборочных средних. Для этого сначала узнаем среднее и дисперсию исходного логнормального распределения.

```python
mean, var = f.stats(moments='mv')
print("Среднее значение исходного распределения: ", mean)
print("Дисперсия исходного распределения: ", var)
```

Согласно ЦПТ распределение выборочных средних аппроксимируется нормальным распределением с тем же средним значением и дисперсией, равной дисперсии исходного распределения, делённого на размер выборки. 

```python
sample_mean = mean
sample_vars = list(map(lambda n: var/n, ns))
print("Среднее значение распределения средних выборочных: ", sample_mean)
print("Дисперсия распределения средних выборочных для n=", ns, ": ", sample_vars)
```

Теперь построим графики распределений выборочных средних для разных значений размера выборки и поверх каждого из графика построим нормальное распределение. 

```python
for i, v in enumerate(sample_means_by_n.items()):
    values = v[1]
    n = v[0]
    plt.hist(values, bins=20, density=True)
    
    # Готовим график нормального распределения
    x = np.linspace(min(values), max(values), 100)
    norm_rv = norm(sample_mean, math.sqrt(sample_vars[i]))
    norm_pdf = norm_rv.pdf(x)
    plt.plot(x, norm_pdf)
    plt.title(n)
    plt.show()
```

**Вывод:** как видно из полученных графиков по мере увеличения размера выборки распределение выборочных средних становится всё ближе к соответствующему нормальному распределению, задаваемому средним значением исходного распределения и дисперсией, обратной пропорциональнйо размеру выборки.
