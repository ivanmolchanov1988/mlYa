```python
import math
import pylab
import numpy as np
from matplotlib import mlab
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.optimize import differential_evolution
```

```python
def func(x):
    return math.sin(x/5)*math.exp(x/10)+5*math.exp(-x/2)
```

```python
# Интервал изменения переменной по оси X
xmin = 1.
xmax = 30.1

# Шаг между точками
dx = 0.1

# Список координат по оси X на отрезке [-xmin; xmax], включая концы*см xmax;
#xlist = mlab.frange(xmin, xmax, dx)
xlist = np.arange(xmin, xmax, dx)
#print(xlist)

# Значение функции в заданных точках
ylist = [func(x) for x in xlist]
#print(ylist)

# Рисуем одномерный график
#pylab.plot(xlist, ylist)
fx = [xlist, ylist]
plt.plot(fx[0],fx[1])
print(min(ylist))
#print(fx[0], fx[1])
```

```python
fmin = minimize(func, [30], method='BFGS')
print(fmin.fun, fmin)
oxmin = np.arange(1, 26, 1)
ymin = [x for x in oxmin]
fminx = [oxmin, ymin]
#print(fminx[1])
plt.plot(fx[0],fx[1] , fmin.x[0],fmin.fun,marker='o')
```

```python
bounds = [[1,30]]
fdifEvolution = differential_evolution(func, bounds)
print(fdifEvolution.fun, fdifEvolution)
plt.plot(fx[0], fx[1] , fdifEvolution.x[0],fdifEvolution.fun, marker='o')
```

```python
def fun(h):
    return int(func(h))
```

```python
# Значение функции в заданных точках
hlist = [fun(h) for h in xlist]
#print(hlist)

# Рисуем одномерный график
#pylab.plot(xlist, ylist)
hx = [xlist, hlist]
plt.plot(hx[0],hx[1])
#print(fx[0], fx[1])
```

```python
fmin2 = minimize(fun, [30], method='BFGS')
print(fmin2.fun, fmin2)
#print(fminx[1])
plt.plot(hx[0],hx[1] , fmin2.x[0],fmin2.fun,marker='o')
```
