```python
import math
import pylab
import numpy as np
from matplotlib import mlab
import matplotlib.pyplot as plt
```

```python
def func(x):
    return math.sin(x/5)*math.exp(x/10)+5*math.exp(-x/2)
```

```python
# Интервал изменения переменной по оси X
xmin = 1.
xmax = 15.1

# Шаг между точками
dx = 0.1

# Список координат по оси X на отрезке [-xmin; xmax], включая концы*см xmax;
#xlist = mlab.frange(xmin, xmax, dx)
xlist = np.arange(xmin, xmax, dx)
#print(xlist)

# Значение функции в заданных точках
ylist = [func(x) for x in xlist]
#print(ylist)

# Рисуем одномерный график
#pylab.plot(xlist, ylist)
fx = [xlist, ylist]
plt.plot(fx[0],fx[1])
#print(fx)
```

```python
xt = 8.

x2t = 4.
x3t = 10.

f1 = func(xmin)
f15 = func(xmax - 0.1)

arr1 = np.array([[1., xmin], [1., xmax-dx]])
arrf1 = np.array([func(xmin), func(xmax-dx)])

arr2 = np.array([[1., xmin, xmin**2],[1., xt, xt**2], [1. ,xmax-dx ,(xmax-dx)**2]])
arrf2 = np.array([func(xmin), func(xt) , func(xmax-dx)])

arr3 = np.array([[1., xmin, xmin**2, xmin**3],[1., x2t, x2t**2, x2t**3],[1., x3t, x3t**2, x3t**3],[1. ,xmax-dx ,(xmax-dx)**2, (xmax-dx)**3]])
arrf3 = np.array([func(xmin), func(x2t), func(x3t), func(xmax-dx)])

w = np.linalg.solve(arr1, arrf1)
w2 = np.linalg.solve(arr2, arrf2)
w3 = np.linalg.solve(arr3, arrf3)

print(arr1, '\n\n', arrf1, '\n\n', w[0], ' ', + w[1])
```

```python
def f1(xOne):
    return w[0] + w[1]*xOne
```

```python
f1List = [f1(xOne) for xOne in xlist]
f1x = [xlist, f1List]

plt.plot(f1x[0],f1x[1] , fx[0],fx[1])
```

```python
def f2(ox):
    return w2[0] + w2[1]*ox + w2[2]*(ox)**2
```

```python
#x2List = [1.,8.,15.]
x2List = np.arange(xmin,xmax,dx)
f2List = [f2(x2) for x2 in x2List]
f2x = [x2List, f2List]

plt.plot(f1x[0],f1x[1] , fx[0],fx[1], f2x[0],f2x[1])
```

```python
def f3(ox3):
    return w3[0] + w3[1]*ox3 + w3[2]*(ox3)**2 + w3[3]*(ox3)**3
```

```python
#x3List = [1.,4.,10.,15.]
x3List = np.arange(xmin,xmax,dx)
f3List = [f3(ox3) for ox3 in x3List]
f3x = [x3List, f3List]

plt.plot(f1x[0],f1x[1] , fx[0],fx[1], f2x[0],f2x[1], f3x[0],f3x[1])

print (round(w3[0],2), '\n', round(w3[1],2), '\n', round(w3[2],2), '\n', round(w3[3],2))
```
