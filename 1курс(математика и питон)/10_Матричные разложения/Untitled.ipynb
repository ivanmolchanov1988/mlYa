```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
mu = 2.0
sigma = 0.5

norm_rv = sts.norm(loc = mu, scale = sigma)

norm_rv.rvs(size = 10)
```

```python
norm_rv.cdf(2.79779891)
```

```python
xlist = norm_rv.rvs(size = 10000)
#print(xlist)
```

```python
ylist = [norm_rv.cdf(x) for x in xlist]
#print(ylist)
```

```python
fx = [xlist, ylist]
plt.plot(fx[0],fx[1])
```

```python
x = np.linspace(0,4,100)
cdf = norm_rv.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
#Построим график функции распределения:
```

```python
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
#Построим график функции плотности:
```

```python
#Равномерное распределение на отрезке
a = 1
b = 4
uniform_rv = sts.uniform(a, b-a)

uniform_rv.rvs(10)
```

```python
x = np.linspace(0,5,100)
cdf = uniform_rv.cdf(x)
plt.plot(x, cdf)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,5,100)
pdf = uniform_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
x = np.linspace(0,30,1000)
for l in [1, 5, 10, 15]:
    rv = sts.poisson(l)
    cdf = rv.cdf(x)
    plt.step(x, cdf, label="$\lambda=%s$" % l)
plt.legend()
plt.title("CDF (poisson)")

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
```

```python
#Пуассон
from collections import Counter
poisson_rv = sts.poisson(3)
c = poisson_rv.rvs(100)
cc = Counter(c)
print(cc)
print("Вероятности выпадений каждой из сторон:")
print({k: v/100.0 for k, v in cc.items()})
```
